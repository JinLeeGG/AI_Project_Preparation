```python
i = 1
while i <= 5:
    print('Hello Python')
    i += 1
print(i)
```

```python
# adding 1 ~ 10
i = 1
sum = 0

while i <= 10:
    sum += 1
    i += 1

print(f'result = {sum}')
```

```python
# adding even number of 1 to 100
i = 1
sum = 0

while i <= 100:
    if i % 2 == 0:
      sum += i
    i += 1

print(f'result {sum}')
```

```python
# print out mulitplication table
dan = int(input('insert the row number: '))
print(f'table {dan}')

i = 1
while i <= 9:
    print(f'{dan} * {i} = {dan * i}')
    i += 1
```

```python
# add even number of 1 to 100:

sum = 0
for i in range(0, 101, 2): # step as 2 for calculating even number
    sum += i

print(f'result {sum}')
```

```python
# ZIP function

li1 = [10, 20, 30]
li2 = ['apple', 'banana', 'orange']

for i in range(len(li1)):
    print(li1[i], li2[i])
```

```python
# if one list is longer than the other:

li1 = [10, 20, 30, 40, 50]
li2 = ['apple', 'banana', 'orange']

for i in range(len(li1)):
    print(li1[i], li2[i]) # index out of range
```

```python
# use zip for this problem

li1 = [10, 20, 30]
li2 = ['apple', 'banana', 'orange']

for l1, l2 in zip(li1, li2):
    print(l1, l2)
```

```python
# it does not cause error if one of list is longer

li1 = [10, 20, 30, 40]
li2 = ['apple', 'banana', 'orange']

for l1, l2 in zip(li1, li2):
    print(l1, l2)
```

```python
li1 = [10, 20, 30, 40]
li2 = ['apple', 'banana', 'orange']

for li in zip(li1,li2):
    print(li) #prints out as tuple
```

```python
# Double looping

for i in range(1, 4):
    print(f'@ i: {i}')
    for j in range(1, 4):
        print(f'  $ i: {j}')
```

```python
for i in range(5):
    for j in range(5):
      print('@', end=' ')
    print()
```

```python
for i in range(5):
    for j in range(5-i):
      print('@', end=' ')
    print()
```

```python
for i in range(5):
    for j in range(i+1):
      print('#', end = ' ')
    print()
```

```python
# generating multiplication table using for loop

for i in range(2, 10):
    print(f'table #: {i}')
    for j in range(1, 10):
      print(f'{i} * {j} = {i * j}')
```

```python
# using while loop
i = 2
j = 1
while i <= 9:
  print(f'table #: {i}')
  while j <= 9:
    print(f'{i} * {j} = {i * j}')
    j += 1
  j = 0
```
