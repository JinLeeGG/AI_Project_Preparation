# 2. Series와 DataFrame

### 1. Series

Series는 1차원 배열과 같은 자료구조로 하나의 열을 나타냅니다. 또한 각 요소는 인덱스(index)와 값(value)으로 구성되어 있습니다. 값은 넘파이의 ndarray 기반으로 저장됩니다. Series는 다양한 데이터 타입을 가질 수 있으며 정수, 실수, 문자열 등 다양한 형태의 데이터를 담을 수 있습니다.

```python
idx = ['김사과', '반하나', '오렌지', '이메론', '배애리']
data = [67, 75, 90, 62, 98]

# pd.Series(데이터, 인덱스, ...)
pd.Series(data)
```

```python
se1 = pd.Series(data, idx)
se1
```

```python
# se1의 index 값을 출력
print(se1.index)
# se1의 value 값을 출력 (ndarray로 저장됨)
print(se1.values)
print(type(se1.values)) # value는 ndarray 이다.
```

### 2. DataFrame

데이터프레임(DataFrame)은 판다스(Pandas) 라이브러리에서 제공하는 중요하고 강력한 데이터 구조로, 2차원의 테이블 형태 데이터를 다루는 데 사용됩니다. 또한 각 요소는 인덱스(index), 열(column), 값(value)으로 구성되어 있습니다. 데이터프레임은 행과 열로 이루어져 있으며, 각 열은 다양한 데이터 타입을 가질 수 있습니다. 값은 넘파이의 ndarray 기반으로 저장됩니다.

```python
data = [[67, 93, 91],
        [75, 68, 96],
        [87, 81, 82],
        [62, 70, 75],
        [98, 56, 87]]

indx = idx = ['김사과', '반하나', '오렌지', '이메론', '배애리']
col = ['국어', '영어', '수학']
```

```python
# pd.DataFrame(데이터, 인덱스, 컬럼, ...)
pd.DataFrame(data)
```

```python
pd.DataFrame(data, idx)
```

```python
pd.DataFrame(data, idx, col)
```

```python
# 매개변수의 이름을 적어주면 순서의 관계가 없다.
df = pd.DataFrame(index=idx, columns=col, data=data)
df
```

```python
print(df.index) # object type (문자가 들어가는건 모두 object. String이 존재하지 않음)
print(df.columns)
print(df.values) # ndarray
```

```python
# 딕셔너리를 사용하여 데이터프레임을 생성하기
dic = {
    '국어':[67, 75, 76, 62, 98],
    '영어':[93, 68, 81, 70, 56],
    '수학':[91, 96, 82, 75, 87]
}

df = pd.DataFrame(data=dic, index=idx)
df
```

```python
# loc 인덱싱: 컬럼 인덱싱
df.loc[:, 'name'] # 전체행, 열은 이름만
# df['name'] or df.name
```

```python
# 2~5까지 가져오기 (끝 번호를 포함한 index 번호이다.)
df.loc[2:5, 'name'] # 5번을 포함
```

```python
# 여러 columns 를 가져올수 있다. (슬라이싱)
df.loc[2:5, ['name', 'gender', 'height']]
```

```python
df.loc[[2,5], ['name', 'gender', 'height']]
```

```python
df.loc[[2,5], 'name':'height']
```

```python
# iloc 인덱싱: index로 인덱싱
df.iloc[:, 0] # 다 포함하되 0번 콜롬만 가져오기
```

```python
df.iloc[:, 0:3] # 3을 포함하지 않음 (이건 범위를 포함하지 않는다)
```

```python
df.iloc[:, [0, 3]] # 0, 3콜롬
```

```python
df.iloc[1:5, 0:2] # 1~5가 되기 직전까지, 0부터 2가되기 직전까지
```

```python
# 모든 데이터를 True False로 나눈다.
df['height'] >= 180 # 키가 180이 넘는사람들
```

```python
df[df['height'] >= 180] # 키가 180이 넘는 사람의 데이터 출력
```

```python
# 키가 180이 넘는 사람의 이름만 보고싶다면
df[df['height'] >= 180]['name'] # df['name'][df['height'] >= 180] (같은 아웃풋)
```

```python
# 180이 넘는 사람의 이름, 성별, 키를 불러와
df[df['height'] >= 180][['name', 'gender', 'height']]
```

```python
# loc 이용해서 데이터 전처리
df.loc[df['height'] >= 180, ['name', 'gender', 'height']]
```

```python
company = ['빅히트', '어도어']
df['company'].isin(company)
```

```python
df[df['company'].isin(company)] # df.loc[df['company'].isin(company),:]
```

# 6. 결측값
결측값은 값이 누락된 데이터를 의미하며, 판다스에서는 일반적으로 NaN으로 표시됩니다.

```python
df.info()
# 19라고 되어있으면 1개씩 빠진거다.
```

```python
df.isna() # 전체데이터에서 nan만 true로 나온다.
```

```python
df.isnull() # NaN을 True
```

```python
df.notnull() # NaN이 False
```

```python
df['height'].isna() #height만 nan
```

```python
df[df['height'].isna()] # height가 nan인것만 뽑기
```

```python
df[df['height'].isna()]['name'] # height가 nan인것중 이름만 뽑기
```

```python
df[df['height'].notnull()] # height값이 nan이 아닌사람만 뽑기
# df[~df['height'].isnull()]  # 같은결과
```

```python
# loc로 df에 company가 nan이 아닌 사람을 찾고 name, company, group, gender 항목만 뽑기
df.loc[df['company'].notnull(), ['name', 'company', 'group', 'gender']]
```
