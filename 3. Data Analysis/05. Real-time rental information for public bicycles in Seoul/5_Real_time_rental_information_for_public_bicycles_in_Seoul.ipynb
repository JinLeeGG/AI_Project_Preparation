```python
import requests
import pandas as pd
import folium
```

```python
# api address
# 맨 마지막 url의 번호를 바꾸어서 데이터의 양을 조절 할 수 있다
# http://openapi.seoul.go.kr:8088/sample/json/bikeList/1/1000/
# http://openapi.seoul.go.kr:8088/sample/json/bikeList/1001/2000/
base_url = 'http://openapi.seoul.go.kr:8088/6b5269637a676a6c3131346177754574/json/bikeList/1/1000/'
response = requests.get(base_url)
response # normal access
```

```python
# 인터넷에서 받은 문자열을 파이썬에서 사용할 수 있는 json형태로 변경
json_data = response.json()
json_data # json 형태가 된다.
```

```python
# rentBikeStatus를 가져오는데, 있으면 가져오고 없으면 {} 반환
json_data.get('rentBikeStatus', {})
```

```python
# rentBikeStatus 안에있는 RESULT를 가져와라
json_data.get('rentBikeStatus', {}).get("RESULT", {})
```

```python
# rentBikeStatus 안에있는 RESULT 안의 CODE를 가져와라
json_data.get('rentBikeStatus', {}).get("RESULT", {}).get("CODE", {})
```

```python
# 원래 dictionary 접근은 이런식으로 했다.
# 하지만 이런식으로 접근 안하는 이유는 만약 에러가 발생했을경우 프로그램이 강종당한다.
# 따라서 위의 방법을 사용한다.
json_data['rentBikeStatus']['RESULT']['CODE']
```

```python
# 데이터를 저장해보자 (약 2000개의 데이터가 있는걸 확인했다.)
def fetch_bike_data():
    base_url = 'http://openapi.seoul.go.kr:8088/6b5269637a676a6c3131346177754574/json/bikeList/'
    start = 1
    end = 1000
    step = 1000
    data_frames = []

    while True:
        # start 부터 시작해서 end 까지 써준다.
        url = f'{base_url}{start}/{end}/'
        # 사이트에 접속
        response = requests.get(url)

        # 만약 정상적으로 접속을 하지 못할경우
        if response.status_code != 200:
            print(f'status code: {response.status_code}')

        # 제이슨화 시켜서 파이썬에서 사용가능하게 만들기
        json_data = response.json()

        # 파싱 시작
        try:
            # 일단 데이터셋 가져오기
            rent_bike_status = json_data['rentBikeStatus']

            # 코드를 가져오기
            result_code = rent_bike_status['RESULT']['CODE']

        # 만약 데이터셋을 못가져올경우
        except:
            print('JSON err')
            break

        # 만약 데이터셋에 데이터가 없을경우
        if result_code == 'INFO-200':
            print("데이터 없음")

        # 데이터가 있을경우
        elif result_code == 'INFO-000':
            print(f'시작: {start} 끝: {end}')
            try:
                # 데이터가 있으면 row를 가져온다.
                bike_data = rent_bike_status['row']
                if bike_data:
                    # DataFrame으로 만들기
                    df = pd.DataFrame(bike_data)
                    # 1000까지 집어넣고, 여기다가 데이터를 계속 쌓기
                    data_frames.append(df)
            except KeyError:
                print("데이터를 찾을 수 없음")
        else:
            print(f'result code: {result_code}')
            break

        # 한 바퀴 다 돌았을 경우 1000을 더해준다.
        # 다음 데이터셋으로 이동하는 과정
        start += step
        end += step

    if data_frames:
        # 데이터 프레임 3개를 합쳐준다. (index를 제외하고 합친다.)
        final_df = pd.concat(data_frames, ignore_index=True)
        return final_df
    else:
        return pd.DataFrame()
```

```python
# 3바퀴를 돌고 데이터가 더 없을경우 제이슨 에러를 일으킨다.
bike_data_df = fetch_bike_data()
```

```python
bike_data_df # 데이터가 잘 생긴걸 보여준다.
```

```python
# 데이터가 non-null로 다 차있다.
# 문자열로부터 가져와서 모든 타입이 object이다.
# Totalcount, 주차총건수, 위도, 경도도 object 이다.
bike_data_df.info()
```


- rackTotCnt	거치대개수
- parkingBikeTotCnt	자전거주차총건수
- shared	거치율
- stationLatitude	위도
- stationLongitude	경도
- stationId	대여소ID
- stationName	대여소이름


```python
bike_data_df.shape
```

```python
# 위도, 경도를 실수로 저장
bike_data_df['stationLongitude'] = bike_data_df['stationLongitude'].astype(float)
bike_data_df['stationLatitude'] = bike_data_df['stationLatitude'].astype(float)
```

```python
bike_data_df.info()
```

```python
# 위도, 경도를 Folium을 사용해서 맵에다가 뿌려주기
# 기준점 위치를 어떻게 보여줄까? 서울에 있으니까? 서울시청?
# 위도의 평균, 경도의 평균을 기준점으로 설정해준다.
bike_map = folium.Map(location=[bike_data_df['stationLatitude'].mean(),
                                bike_data_df['stationLongitude'].mean()],
                      zoom_start=12)

# index와 value를 넘겨주는 enumerate()와 같다.
# 각 행을 뽑아서 넘겨준다.
for index, data in bike_data_df.iterrows():
    # 자전거 역 이름, 총 건수를 찍어준다.
    popup_str = f"{data['stationName']} / 자전거 주차 총 건수: {data['parkingBikeTotCnt']}대"

    # 체크
    # print(popup_str)

    # 팝업(마커에 들어가는 문자)을 폴리움에 등록
    popup = folium.Popup(popup_str, max_width=600)
    # 각 데이터의 위경도를 마커로 찍어준다
    folium.Marker(location=[data['stationLatitude'], data['stationLongitude']],
                  popup=popup).add_to(bike_map)

# 지도 표출
bike_map
```
