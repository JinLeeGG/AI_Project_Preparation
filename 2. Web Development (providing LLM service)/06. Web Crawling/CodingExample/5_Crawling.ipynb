# 1. basicenglishspeacking.com
[Website](https://basicenglishspeaking.com/daily-english-conversation-topics/#google_vignette)

```python
# 내가 원하는 서버에 접속 후 페이지를 가져오게 하는 모듈 (html)
import requests
# 가져온 html, css, javascript 들 중 원하는 일부를 가져와서 가공할 수 있게 도와주는 모듈
from bs4 import BeautifulSoup
```

```python
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
# url를 통해서 접속 후 문서를 가져올수 있는지 없는지가 저장됨
request = requests.get(site)

# response 200 - 정상 접속
# 4XX 접속 불가
print(request)
```

```python
# html, css, javascripts 를 그대로 가져옴
request.text
```

```python
# 전달받은 text를 넣어준다. (객체생성)
soup = BeautifulSoup(request.text)
```

```python
# find로 tag이름, 속성을 넣어서 문서로 만들어진 명령어를 찾는다.
div = soup.find('div', {'class' : 'thrv_wrapper thrv-columns'})
print(div)
```

```python
# hyperlink만 가져오기

# anchor tag만 가져오기
links = div.find_all('a')

print(links)
```

```python
# text만 뽑아오기
for link in links:
  print(link.text)
```

```python
subject = []

for link in links:
    subject.append(link.text)
```

```python
print(subject)
print(len(subject))
```

```python
print(f'총 {len(subject)}개의 주제를 찾았습니다.')
for i in range(len(subject)):
  print(f'{i+1}. {subject[i]}')
```

# 2. 다음 뉴스기사 제목 크롤링
* https://v.daum.net/v/20250609160147728
* https://v.daum.net/v/20250609180414605
* https://v.daum.net/v/20250609180818746

```python
def daum_news_title_crawling(news_id):
    url = f'https://v.daum.net/v/{news_id}'
    request = requests.get(url)
    soup = BeautifulSoup(request.text)
    title = soup.find('h3', {'class':'tit_view'})
    if title:
        return title.text.strip()
    return '제목없음'
```

```python
daum_news_title_crawling(20250609160147728)
```

```python
daum_news_title_crawling(20250609180414605)
```

```python
daum_news_title_crawling(20250609180818746)
```

# 3. 멜론 차트 1~100위까지 크롤링
[사이트](https://www.melon.com/chart/index.htm)

```python
# 처음에는 막힌다 (user agent가 아니기때문).
request = requests.get('https://www.melon.com/chart/index.htm')
request
```

```python
# User Agent 추가
# Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
request = requests.get('https://www.melon.com/chart/index.htm', headers=header)
print(request)
print(request.text)
```

```python
soup = BeautifulSoup(request.text)
```

```python
# title 가져오기
title = soup.findAll('div', {'class':'rank01'})
print(title)
```

```python
# 아티스트 가져오기
artist = soup.findAll('span', {'class':'checkEllipsis'})
print(artist)
```

```python
for i, (t, a) in enumerate(zip(title, artist),1):
    title = t.text.strip()
    artist = a.text.strip()
    print(f'{i}, {title} - {artist}')
```
