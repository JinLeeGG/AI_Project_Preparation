# problem 4
https://school.programmers.co.kr/learn/courses/30/lessons/181859

image.png

```python
def solution(arr):
    stk = []
    i = 0
    while i < len(arr):
        if not stk:
            stk.append(arr[i])
            i += 1
        elif stk[-1] == arr[i]:
            stk.pop()
            i += 1
        else:
            stk.append(arr[i])
            i += 1

    # edge case
    if not stk:
        return [-1]
    return stk
```

```python
print(solution([0, 1, 1, 1, 0]))
print(solution([0, 1, 0, 1, 0]))
print(solution([0, 1, 1, 0]))
```
